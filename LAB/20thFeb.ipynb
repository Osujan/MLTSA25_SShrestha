```python
# prompt: unig the dataframe consumption_df fit a line to consumption vs year with sklearn  include fitting in intercept and plot the result

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

consumption_df = pd.read_csv("https://raw.githubusercontent.com/fedhere/MLTSA_FBianco/refs/heads/main/data/consumptionUS.csv")

# Prepare the data
X = consumption_df[['year']]
y = consumption_df['consumption']

# Create and fit the linear regression model
model = LinearRegression(fit_intercept=True)  # Include the intercept
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Actual Consumption')
plt.plot(X, y_pred, color='red', label='Fitted Line')
plt.xlabel('Year')
plt.ylabel('Consumption')
plt.title('Consumption vs. Year with Linear Regression')
plt.legend()
plt.show()

# Print the model parameters
print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_[0])
```

```python
# prompt: fit a multilinear modle with sklearn to consumption_df prediction consumption, use all variables 'year', 'income',  'production', 'unemployment',
#        'savings' and plot the true value of consu,ption and the prediction over time, also show the sum of squared errors

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Load the dataframe.
consumption_df = pd.read_csv("https://raw.githubusercontent.com/fedhere/MLTSA_FBianco/refs/heads/main/data/consumptionUS.csv")

# Prepare the data
X = consumption_df[['year', 'income', 'production', 'unemployment', 'savings']]
y = consumption_df['consumption']

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Calculate the sum of squared errors
sse = np.sum((y - y_pred)**2)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(consumption_df['year'], y, label='Actual Consumption')
plt.plot(consumption_df['year'], y_pred, color='red', label='Predicted Consumption')
plt.xlabel('Year')
plt.ylabel('Consumption')
plt.title('Actual vs. Predicted Consumption')
plt.legend()
plt.show()

print(f"Sum of Squared Errors: {sse}")
```
