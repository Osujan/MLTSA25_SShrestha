```python
import pandas as pd
```

```python
movies=pd.read_csv("https://raw.githubusercontent.com/fedhere/PyBOOT/refs/heads/master/data/movies.dat", delimiter="::", index_col=0, names = ["name", "gener"])
```

```python
int(movies.loc[1,"name"].split("(")[1].strip(")"))
```

```python
def getyear(name):
  return int(name.split("(")[-1].strip(")"))
movies["year"]=movies["name"].map(getyear)
```

```python
movies
```

```python
moviesbyyear=movies.groupby("year").count()
```

```python
moviesbyyear.plot(y="name")
```
